```python
import serial
import pyautogui as pg
import time
import threading

ser = serial.Serial(port='COM3',baudrate=9600,
        bytesize=serial.EIGHTBITS, parity=serial.PARITY_NONE, timeout=0.05)

try:                     
    ser.isOpen()
    print("Serial port is Open")          
except:
    print("Error")           
    exit() 
try:
    if(ser.isOpen()):
        pg.FAILSAFE=False;
        while(1):
            line=ser.read(1).decode()
            if len(line)!=0:

                if(line=="1"):
                    print('up')                
                    pg.press('up')

                if(line=="2"):
                    print('down')
                    pg.press('down')
                if(line=='4'):
                    print('left')
                    pg.press('left')
                if(line=='3'):
                    print('right')
                    pg.press('right')
               # if(line=='0'):
                    #pg.keyUp('right')
                    #pg.keyUp('left')    
                    #pg.keyUp('up')    
                    #pg.keyUp('down')
                #    print('idle')
                if line=='9':
                    pg.press(['space','space'])  
                ser.flush()
    else:
        print("Cannot Open Serial Port")
except:
    print("Exception or Intrupt Serial closing....")
    ser.close()
```
